## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Data\NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.describe()
```

```python
x = shot['location_x']
y = shot['location_y']
plt.scatter(x,y, s =.1, c = 'r', marker = '.')
```

```python
plt.figure(figsize = (90/6, 50/6))
plt.scatter(x,y, s =.1, c = 'r', marker = '.')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='red')
```

```python
#The right hand half court

plt.figure(figsize = (94/12, 50/6))
plt.scatter(x,y, c = 'r', marker = ".")
plt.minorticks_on()
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(933/2, 933)
```

```python
#The left hand half court

plt.figure(figsize = (94/12, 50/6))
plt.scatter(x,y, c = 'r', marker = ".")
plt.minorticks_on()
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(933, 933/2)
```

```python
#Defining left half courts
shot['halfcourt_x'] =np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] =np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
shot.describe()
```

```python
#The left hand half court
x = shot['halfcourt_x']
y = shot['halfcourt_y']
plt.figure(figsize = (94/12, 50/6))
plt.scatter(x,y, c = 'r', marker = ".")
plt.minorticks_on()
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(933/2, 933)
```

```python
#Defing right half court
shot['halfcourt_x1'] =np.where(shot['location_x'] > 933/2, shot['location_x'] - 900,shot['location_x'])
shot['halfcourt_y1'] =np.where(shot['location_x'] > 933/2, shot['location_y'] - 500,shot['location_y'])
shot.describe()
```

```python
#Right hand half of the court
x = shot['halfcourt_x1']
y = shot['halfcourt_y1']
plt.figure(figsize = (94/12, 50/6))
plt.scatter(x,y, c = 'r', marker = '.')
plt.minorticks_on()
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(0,500)
plt.ylim(-10, 510)
```

```python
f = plt.figure(figsize = (94/6, 50/6))
ax = f.add_subplot(121)
ax = plt.scatter(shot['halfcourt_x'],shot['halfcourt_y'], s = 10, c = 'r', marker = ".")
plt.grid(True)
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(933/2, 933)
ax = f.add_subplot(122)
ax = plt.scatter(shot['halfcourt_x1'], shot['halfcourt_y1'], s = 10, c = 'b', marker = '.')
plt.grid(True)
plt.grid(which = 'major', linewidth = 0.5, color = 'black')
plt.grid(which = 'major', linewidth = 0.5, color = 'red')
plt.xlim(0,500)
plt.ylim(-10, 510)
```

## Westbrook - Durant Comparision

```python
westbrook  = shot[shot['shoot_player'] == 'Russell Westbrook']
durant = shot[shot['shoot_player'] == 'Kevin Durant']
```

```python
hxw = westbrook['halfcourt_x']
hyw = westbrook['halfcourt_y']
colors = np.where(westbrook['current_shot_outcome'] == 'SCORED', 'r', 
                  np.where(westbrook['current_shot_outcome'] == 'MISSED', 'b', 'g'))
plt.figure(figsize = (94/12, 50/6))
plt.scatter(hxw, hyw, c = colors, s = 10, marker = '.')
plt.grid(True)
plt.xlim(500, 950)
plt.ylim(0, 500)
plt.title("Russel Westbrook")
```

```python
hxw = westbrook['location_x']
hyw = westbrook['location_y']
colors = np.where(westbrook['current_shot_outcome'] == 'SCORED', 'r', 
                  np.where(westbrook['current_shot_outcome'] == 'MISSED', 'b', 'g'))
plt.figure(figsize = (94/12, 50/6))
plt.scatter(hxw, hyw, c = colors, s = 10, marker = '.')
plt.grid(True)
plt.title("Russel Westbrook")
plt.savefig("Russel Westbrook.png")
```
